```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.arange(0,21,1)
z = np.empty(21)

for i, x_i in enumerate(x):
    if x_i == 0:
        y = 0.1
    elif x_i == 1:
        y = 0.5
    elif x_i == 2:
        y = 0.9
    else:
        y = 0.9 * np.exp((2-x_i)*3.0)
    z[i] = y


plt.plot(x,z)
plt.xlabel('Training iteration')
plt.ylabel('Probability selecting rating query')
plt.show()
```
